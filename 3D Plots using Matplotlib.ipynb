Surface plots can be great for visualising the relationships among 3 variables across the entire 3D landscape.
They give a full structure and view as to 
how the value of each variable changes across the axes of the 2 others.


### 3D Contour Plot 

```python
def f(x , y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X ,Y)  # Z data will be evaluated from x-data and y-data
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour(X,Y,Z, 50, cmap='hsv') # try both hsv and binary in cmap
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(20,10); # 20 degrees above x-y plane and (elevation) 
                    #and 10 degrees rotated clockwise about z-axis
                    # this is optional
```

### Wireframes

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z , color='black')
ax.set_title('Wireframe');
```

```python
# wireframe another example with diffrent data


ax = plt.axes(projection='3d')
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = R

ax.plot_wireframe(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=3, )
plt.show()
```

## Surface Plot 
A surface plot is like a wireframe plot, but each face of the wireframe is a filled polygon. Adding a colormap to the filled polygons can aid perception of the topology of the surface being visualized:

```python
ax = plt.axes(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.show()

```

```python
ax = plt.axes(projection='3d')
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = R     # here i changed the Z function to see diffrence

ax.plot_surface(X, Y, Z, cmap=cm.BrBG,
                       linewidth=3, )
plt.show()
```
